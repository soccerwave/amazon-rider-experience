```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/processed/cleaned_amazon.csv')

# what exactly does very rider?
df_riders = df.groupby('Agent_Age').agg({
    'Pickup_Delay_Minutes': 'mean',
    'Delivery_Speed_KMPH': 'mean',
    'Distance_KM': 'mean',
    'Agent_Rating': 'mean'
}).reset_index()

```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

features = ['Pickup_Delay_Minutes', 'Delivery_Speed_KMPH', 'Distance_KM', 'Agent_Rating']
X = df_riders[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=4, random_state=7)
df_riders['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df_riders, hue='Cluster', palette='Set2')
plt.suptitle('Rider Profiles by Cluster', y=1.02)
plt.savefig('../reports/screenshots/rider_clusters.png')
plt.show()
```

```python
### Key Insights:

# 1. Cluster 1 riders are generally younger and have the highest delivery speed,
# suggesting they are likely more agile and efficient in last-mile logistics.

#2. Cluster 0 shows the lowest pickup delay and highest agent ratings,
# indicating they represent the most reliable group, likely ideal candidates for rewards.

#3. Cluster 3 maintains a high average delivery distance,
# which could mean they accept longer distance deliveries more frequently which is valuable for logistical coverage.

#4. Cluster 2 includes riders with relatively low ratings and higher pickup delays
# potentially highlighting training or support needs for this group.

#5. There is a clear inverse relationship between delivery speed and pickup delay across clusters,
# implying operational bottlenecks occur more in the pickup phase than during transit.
```
