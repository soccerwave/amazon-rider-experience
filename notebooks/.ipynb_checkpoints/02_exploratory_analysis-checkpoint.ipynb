```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

pd.set_option('display.max_columns', None)

df = pd.read_csv('../data/processed/cleaned_amazon.csv')
df.shape
```

```python

# table of KPIs
kpis = {
    'avg_pickup_delay_min': df['Pickup_Delay_Minutes'].mean(),
    'median_delivery_speed_kmph': df['Delivery_Speed_KMPH'].median(),
    'on_time_delivery_rate': (df['Delivery_Time'] <= 45).mean(),
    'avg_agent_rating': df['Agent_Rating'].mean()
}
kpi_df = pd.DataFrame.from_dict(kpis, orient='index', columns=['value'])
display(kpi_df)
```

```python
# visuals
# agent delay
plt.figure(figsize=(8,4))
sns.histplot(df['Pickup_Delay_Minutes'], bins=40, kde=True)
plt.axvline(df['Pickup_Delay_Minutes'].median(), ls='--')
plt.title('Distribution of Pickup Delay (Minutes)')
plt.xlabel('Minutes'); plt.ylabel('Count')
plt.savefig('../reports/screenshots/pickup_delay_distribution.png', bbox_inches='tight')
plt.show()
```

```python
###  Insights:

#- Most pickups are clustered around 5, 10, and 15 minutes, indicating possible system-defined pickup scheduling.
#- The distribution has distinct peaks, suggesting delays may be rounded or system-driven.
#- Median pickup delay is approximately 10 minutes, which is generally acceptable operationally.
```

```python
# speed by vehicle type
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='Vehicle', y='Delivery_Speed_KMPH')
plt.ylim(0, df['Delivery_Speed_KMPH'].quantile(0.95))  #remove wrong results
plt.title('Delivery Speed by Vehicle Type')
plt.savefig('../reports/screenshots/speed_by_vehicle.png', bbox_inches='tight')
plt.show()
```

```python
### Insights:

#- Average delivery speed is quite similar across vehicle types.
#- Scooters and vans have slightly higher median speeds than motorcycles.
#- Motorcycles show wider variance, possibly due to more diverse route lengths or rider behavior.
#- Presence of outliers (above 12–14 km/h) may indicate exceptional cases or GPS errors.
```

```python
# effect of traffic on delivery time
plt.figure(figsize=(8,4))
sns.barplot(data=df, x='Traffic', y='Delivery_Time', errorbar=('ci', False))
plt.title('Avg Delivery Time by Traffic')
plt.ylabel('Minutes')
plt.savefig('../reports/screenshots/delivery_time_by_traffic.png', bbox_inches='tight')
plt.show()
```

```python
### Insights:

#- Delivery time is shortest under "Low" traffic conditions (~100 mins).
#- Under "Jam", delivery time increases significantly to around 150 mins (~50% higher than Low).
#- Optimization strategies should consider rerouting during high-traffic periods to improve efficiency.
```

```python
# Correlation Table
plt.figure(figsize=(10,6))
sns.heatmap(df[['Pickup_Delay_Minutes','Delivery_Time','Delivery_Speed_KMPH',
                'Distance_KM','Agent_Rating']].corr(), annot=True, fmt=".2f")
plt.title('Correlation Matrix')
plt.savefig('../reports/screenshots/correlation_matrix.png', bbox_inches='tight')
plt.show()
```

```python
### 🔍 Insights:

#- Delivery speed is negatively correlated with delivery time (-0.45), as expected.
#- Delivery speed shows a moderate positive correlation with delivery distance (+0.44), indicating that longer routes tend to be faster.
#- Agent rating has weak correlation with most operational metrics, suggesting customer satisfaction is influenced by external or subjective factors.
#- Pickup delay has almost no correlation with delivery speed or delivery time, meaning other factors might dominate its variance.
```
